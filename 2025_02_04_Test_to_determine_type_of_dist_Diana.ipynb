## Shapiro-Wilk Test

The Shapiro-Wilk test is a statistical procedure for testing whether a sample of data comes from a normally distributed population. It calculates a W statistic, which ranges from 0 to 1. A W value close to 1 indicates that the sample is likely drawn from a normal distribution, while smaller values suggest deviation from normality.

## Anderson-Darling Test

The Anderson-Darling test is used to determine if a sample of data is drawn from a specific probability distribution. It's particularly effective at detecting departures from normality in the tails of the distribution. The test calculates an A² statistic, with smaller values indicating a better fit to the specified distribution.

## Kolmogorov-Smirnov Test

The Kolmogorov-Smirnov test is a nonparametric test that compares a sample with a reference probability distribution or compares two samples. It quantifies the distance between the empirical distribution function of the sample and the cumulative distribution function of the reference distribution. The test produces a D statistic, which represents the maximum distance between these two functions.


## Q-Q Test

The Q-Q test, or Quantile-Quantile plot, is a graphical method used to determine if a set of data potentially comes from a theoretical distribution, most commonly the normal distribution.

- If the data is normally distributed, the points in a Q-Q plot will lie on a straight diagonal line.
- Deviations from this straight line indicate departures from normality.

Q-Q plots are created by plotting the quantiles of the observed data against the quantiles of the theoretical distribution. This method provides visual feedback on the nature of the data and is considered more robust than simple normality tests.

## Skewness Test

Skewness is a measure of the asymmetry of a probability distribution about its mean. The skewness test assesses the degree and direction of asymmetry in a dataset.

- A skewness value of 0 indicates perfect symmetry (although this is rare in real-world data).
- Positive skewness indicates a distribution with an asymmetric tail extending toward more positive values.
- Negative skewness indicates a distribution with an asymmetric tail extending toward more negative values.

Interpreting skewness values:
- If skewness is less than -1 or greater than 1, the distribution is highly skewed.
- If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.
- If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.

The formula for calculating skewness is:

skewness = (3 * (mean - median)) / standard deviation


Both the Q-Q plot and skewness test provide valuable insights into the normality and distribution characteristics of a dataset, complementing other normality tests in statistical analysis.


## Median

The median is the middle value in a sorted set of numbers, separating the higher half from the lower half of a data sample. It represents the 50th percentile, where half of the values fall above and half below. The median is particularly useful when dealing with skewed data, as it's not affected by extreme values.

## Mean

The mean, often referred to as the average, is calculated by summing all values in a dataset and dividing by the number of values. Unlike the median, the mean can be influenced by outliers or extreme values in the dataset.

## Standard Deviation (std)

Standard deviation is a measure of the amount of variation or dispersion in a set of values. It's calculated as the square root of the variance:

1. A low standard deviation indicates that values tend to be close to the mean.
2. A high standard deviation indicates that values are spread out over a wider range.

The formula for standard deviation is:

s = √[Σ(x_i - x̄)² / (n - 1)]


Where:
- `s` is the sample standard deviation
- `x_i` are the individual values
- `x̄` is the mean
- `n` is the number of values

## Probability Density Function (PDF)

The probability density function (PDF) is a function that describes the relative likelihood for a continuous random variable to take on a given value. Key properties of a PDF include:

1. It's non-negative for all values of the random variable.
2. The total area under the PDF curve equals 1.

The PDF is used to calculate the probability that a random variable falls within a specific range:

P(a < X ≤ b) = ∫[a to b] f_X(x) dx


Where `f_X(x)` is the PDF of the random variable `X`.

These statistical concepts are fundamental in data analysis, providing different ways to understand and describe datasets, their central tendencies, variability, and probability distributions.

```python
#Testing if the data is weibull

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, shapiro, weibull_min, anderson
import scipy.stats as st


plt.hist(array, bins=15, color='purple', density=True)

mean = array.mean()
std = array.std()

x = np.linspace(min(array), max(array), 100)

y = norm.pdf(x, mean, std)

plt.plot(x,y)


#check if the data follows a Weibull distribution


#fit the data to a Weibull distribution
params = weibull_min.fit(array, floc=0)

#perform the Kolmogorov-Smirnov test
ks_stat, ks_p_value = st.kstest(array, 'weibull_min', args=params)

print(f"KS test statistic: {ks_stat}, p-value: {ks_p_value}")

if ks_p_value > 0.05:
    print("The data follows a Weibull distribution")
else:
    print("The data does not follow a Weibull distribution")
```
